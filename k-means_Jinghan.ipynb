```
dfdf = [column for column in sparkdf.columns if column.endswith("0_3m_spend")]
df1 = sparkdf.select(dfdf)
display(df1)
```

```
df1.columns
```

```
data_clean=df1.na.drop()
```

```
from pyspark.ml.feature import VectorAssembler
data_clean.columns
```

```
assemble=VectorAssembler(inputCols=[
 'imp_c_automotive_0_3m_spend',
 'imp_c_communications_service_providers_0_3m_spend',
 'imp_c_entertainment_0_3m_spend',
 'imp_c_other_retail_0_3m_spend',
 'imp_c_restaurant_0_3m_spend',
 'imp_sc_apparel_0_3m_spend',
 'imp_sc_casual_dining_0_3m_spend',
 'imp_sc_convenience_stores_0_3m_spend',
 'imp_sc_drug_stores_0_3m_spend',
 'imp_sc_electronics_0_3m_spend',
 'imp_sc_fast_casual_0_3m_spend',
 'imp_sc_home_improvement_0_3m_spend',
 'imp_sc_mass_merchandiser_0_3m_spend',
 'imp_sc_quick_serve_0_3m_spend',
 'imp_sc_supermarkets_0_3m_spend',
 'imp_sc_warehouse_clubs_0_3m_spend'], outputCol='features')
assembled_data=assemble.transform(data_clean)
assembled_data.show(2)
```

```
from pyspark.ml.feature import StandardScaler
scale=StandardScaler(inputCol='features',outputCol='standardized')
data_scale=scale.fit(assembled_data)
data_scale_output=data_scale.transform(assembled_data)
data_scale_output.show(2)
```

```
from pyspark.ml.clustering import KMeans
from pyspark.ml.evaluation import ClusteringEvaluator
silhouette_score=[]
evaluator = ClusteringEvaluator(predictionCol='prediction', featuresCol='standardized', \
                                metricName='silhouette', distanceMeasure='squaredEuclidean')
for i in range(2,10):
    
    KMeans_algo=KMeans(featuresCol='standardized', k=i)
    
    KMeans_fit=KMeans_algo.fit(data_scale_output)
    
    output=KMeans_fit.transform(data_scale_output)
    
    
    
    score=evaluator.evaluate(output)
    
    silhouette_score.append(score)
    
    print("Silhouette Score:",score)
```

```
#Visualizing the silhouette scores in a plot
import matplotlib.pyplot as plt
fig, ax = plt.subplots(1,1, figsize =(8,6))
ax.plot(range(2,10),silhouette_score)
ax.set_xlabel('cluster')
ax.set_ylabel('cost')
```

```
k = 8
kmeans = KMeans().setK(k).setSeed(1).setFeaturesCol("features")
model = kmeans.fit(data_scale_output)
centers = model.clusterCenters()

print("Cluster Centers: ")
for center in centers:
    print(center)
```

```
KMeans_=KMeans(featuresCol='features', k=8) 
KMeans_Model=KMeans_.fit(assembled_data)
KMeans_Assignments=KMeans_Model.transform(assembled_data)
```

```
data_array =  np.array(df1.select('c_automotive_0_3m_spend',
 'c_communications_service_providers_0_3m_spend',
 'c_entertainment_0_3m_spend',
 'c_other_retail_0_3m_spend',
 'c_restaurant_0_3m_spend',
 'sc_apparel_0_3m_spend',
 'sc_casual_dining_0_3m_spend',
 'sc_convenience_stores_0_3m_spend',
 'sc_drug_stores_0_3m_spend',
 'sc_electronics_0_3m_spend',
 'sc_fast_casual_0_3m_spend',
 'sc_home_improvement_0_3m_spend',
 'sc_mass_merchandiser_0_3m_spend',
 'sc_quick_serve_0_3m_spend',
 'sc_supermarkets_0_3m_spend',
 'sc_warehouse_clubs_0_3m_spend').collect())
dataset = data_array
```

```
within_cluster_variance = []
for k in range(2, 20):
    kmeans = KMeans(n_clusters=k, n_init=20)
    kmeans.fit(dataset)
    within_cluster_variance.append(kmeans.inertia_)
fig = plt.figure(figsize=(15, 5))                                                          
plt.plot(range(2, 20), within_cluster_variance, marker='o')
plt.grid(True)
plt.xlabel('Number of clusters')
plt.ylabel('Reduction in within cluster variation')
plt.title('Elbow curve')                                                                                                                               
```
